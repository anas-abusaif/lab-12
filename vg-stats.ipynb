```python
import pandas as pd
data_frame = pd.read_csv('./vgsales.csv')
data_frame
```

```python
# most common Publisher
most_common_publisher=data_frame['Publisher'].mode()[0]
```

```python
# most common platform
most_common_platform=data_frame['Platform'].mode()[0]
```

```python
# most common genre
most_common_genre=data_frame['Genre'].mode()[0]
```

```python
# top 20 highest grossing games
top_twenty_highest_grossing_games=data_frame[['Rank','Name']].head(20)
top_twenty_highest_grossing_games
```

```python
# median for north american sales
na_median_sales=data_frame['NA_Sales'].median()
```

```python
# ten games surrounding the median sales output
five_above=data_frame[data_frame.NA_Sales>na_median_sales].tail(5)
five_below=data_frame[data_frame.NA_Sales<na_median_sales].head(5)
ten_median_na_seller_names=pd.concat([five_above,five_below])
ten_median_na_seller_names
```
